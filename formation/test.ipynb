```python
import allotment_matching_1 as am
import networkx as nx
import numpy as np
```

```python
num_pts = 1000
source_pts = am.generate_source_pts(num_pts, [0, 0], [1000, 1000])
target_pts = am.generate_source_pts(num_pts, [0, 0], [1000, 1000])

dist_list = []
for s in source_pts:
    for t in target_pts:
        dist_list.append(np.linalg.norm(s - t))
```

```python
g = nx.Graph()
source_nodes = []
for s in range(len(source_pts)):
    a = 's_{}'.format(s)
    source_nodes.append(a)
    g.add_node(a)
    g.nodes[a]['x'] = float(source_pts[s][0])
    g.nodes[a]['y'] = float(source_pts[s][1])
        # g.nodes[a]['l'] = 0
target_nodes = []
for t in range(len(target_pts)):
    b = 't_{}'.format(t)
    target_nodes.append(b)
    g.add_node(b)
    g.nodes[b]['x'] = float(target_pts[t][0])
    g.nodes[b]['y'] = float(target_pts[t][1])

for i, s in enumerate(source_nodes):
    for j, t in enumerate(target_nodes):
        g.add_edge(s, t)
        g[s][t]['length'] = dist_list[i * len(source_pts) + j]
```

```python
nx.bipartite.minimum_weight_full_matching(g, source_nodes, weight = 'length')
```
